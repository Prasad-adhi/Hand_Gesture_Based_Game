```python
import numpy as np
import pandas as pd
import random
```

```python
columns=['X_coor','Y_coor','left']
```

```python
df=pd.DataFrame(columns=columns)
```

```python
random.seed(200)
#random.seed(400)
```

```python
no_of_rows=2000
```

```python
#left=[True,False]
left=[1,0]
```

```python
for i in range(no_of_rows):
    temp=[]
    temp.append(random.randint(0,77))
    temp.append(random.randint(20,120))
    temp.append(left[random.randint(0,1)])
    df.loc[i]=temp
```

```python
df
```

```python
def game(X_coor,Y_coor,left):
    done = False
    grid=np.zeros([80,140],dtype=int)
    grid[:,0]=100
    grid[:,138]=-100
    grid[0,:]=2
    grid[78,:]=2
    grid=np.transpose(grid)
    grid[20,:]=-1
    grid[120,10:22]=1
    grid[Y_coor,X_coor]=5
    pug_x=np.where(grid==5)[1][0]
    pug_y=np.where(grid==5)[0][0]
    pug_top=True
    pug_left=left
    x=np.where(grid==1)[1][0]
    y=np.where(grid==1)[0][0]
    color = (255, 100, 0)
    while not done:
        if(pug_top):
            if(grid[pug_y-1,pug_x]==-1):
                return pug_x
            else:
                grid[pug_y,pug_x]=0
                pug_y-=1
                grid[pug_y,pug_x]=5
        
        if(pug_left):
            if(grid[pug_y,pug_x-1]==2):
                pug_left=False
            else:
                grid[pug_y,pug_x]=0
                pug_x-=1
                grid[pug_y,pug_x]=5
        else:
            if(grid[pug_y,pug_x+1]==2):
                pug_left=True
            else:
                grid[pug_y,pug_x]=0
                pug_x+=1
                grid[pug_y,pug_x]=5
```

```python
labels=[]
for i in range(no_of_rows):
    labels.append(game(df.iloc[i][0],df.iloc[i][1],df.iloc[i][2]))
```

```python
df['labels']=labels
```
